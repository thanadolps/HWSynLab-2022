```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.transform import resize
```

```python
# Load

img = plt.imread('bg.webp')
print(img.shape)
plt.imshow(img)
plt.show()
```

```python

# Resize

VGA_shape = (480, 640)

img = resize(img, tuple(i/2 for i in VGA_shape))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
# Quantization

result = np.digitize(img, np.linspace(0, 1, 16))-1
plt.imshow(result/15)
plt.show()
```

```python
mem = ""

for row in result:
    for pixel in row:
        mem += "".join(hex(x)[2:] for x in pixel)
        mem += " "
    mem += "\n"
```

```python
with open("output.mem", 'w') as file:
    file.write(mem)
```
